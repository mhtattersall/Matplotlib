```python
# !pip install matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
# matplotlib.pyplot is a collection of functions that make matplotlib work like MATLAB
# each pyplot function makes some change to a figure
# all sequences are converted to numpy arrays internally
```

```python
# Components
# Almost every element of a chart can be customised as a python object.
# Figure object is the outermost container of matplotlib graphic, which can contain multiple axes objects.
# Below axes are the actual y-axis and y-axis, individual lines, title, legends, text boxes, ticks, tick labels, grids, spines
```

```python
# Example:
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.axis((0, 5, 0, 20)) #[xmin, xmax, ymin, ymax]
plt.ylabel('some numbers')
plt.show()
```

```python
# For every x,y pair of arguments, there is an optional third argument which is the format string that indicates the color and line type of the plot
# The default format string is 'b-', which is a solid blue line
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'ro') #ro is red circles
plt.show()
```

```python
# Example: (simple sine wave)
x = np.arange(0,10,0.1)
y = np.sin(x)
plt.plot(x,y)
```

```python
# Example: (sine wave from -10 to 10)
x = np.linspace(-10,10,1000)
y = np.sin(x)
plt.figure(figsize = (12,8))
plt.plot(x, y, label='Sine Wave')
plt.title("f(x) = sin(x)", fontsize=20)
plt.xlabel("x", fontsize=15)
plt.ylabel("y", fontsize=15)
plt.title('Sine Wave')
plt.legend()
plt.show()
```

```python
# Plotting categorical variables
names = ['group_a', 'group_b', 'group_c']
values = [1, 10, 100]
plt.figure(figsize=(9, 3))
plt.subplot(131) #(131) is creates a 1row x 3cols grid of subplots, and it's selecting the first subplot (index 1) to be the current subplot.
plt.bar(names, values)
plt.subplot(132)
plt.scatter(names, values)
plt.subplot(133)
plt.plot(names, values)
plt.suptitle('Categorical Plotting')
plt.show()
```

```python
# Line properties
# Lines have many attributes that you can set: linewidth, dash style, antialiased, etc; see matplotlib.lines.Line2D
# Use keyword arguments:
# plt.plot(x, y, linewidth=2.0)
# or use MATLAB style string value pairs
# plt.setp(lines, 'color', 'r', 'linewidth', 2.0)
# or use setter methods
# line, = plt.plot(x, y, '-')
# line.set_antialiased(False) # turn off antialiasing
# to get a list of settable line properties, call the setp function with a line or lines as argument
lines = plt.plot([1, 2, 3])
plt.setp(lines)
```

```python
# Working with text
# xlabel, ylabel and title are used to add text in the indicated locations
# 'text' is can be used to add text in an arbitrary location
# Example:
mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)
# the histogram of the data
n, bins, patches = plt.hist(x, 50, density=True, facecolor='g', alpha=0.75)
plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title('Histogram of IQ')
plt.text(60, .025, r'$\mu=100,\ \sigma=15$') #Writing mathemaatical expressions in text
plt.axis([40, 160, 0, 0.03])
plt.grid(True) # adds gridlines
plt.show()
```

```python
# Annotating text
# To annotate some feature of the plot, state the location being annotated (xy) and the location of the text (xytext)
# Example:
ax = plt.subplot()
t = np.arange(0.0, 5.0, 0.01)
s = np.cos(2*np.pi*t)
line, = plt.plot(t, s, lw=2)
plt.annotate('local max', xy=(2, 1), xytext=(3, 1.5),
             arrowprops=dict(facecolor='black', shrink=0.05),
             )
plt.ylim(-2, 2)
plt.show()
```

```python
# Logarithmic and other nonlinear axes
# An example of four plots with the same data and different scales for the y-axis is shown below
# Fixing random state for reproducibility
np.random.seed(19680801)

# make up some data in the open interval (0, 1)
y = np.random.normal(loc=0.5, scale=0.4, size=1000)
y = y[(y > 0) & (y < 1)]
y.sort()
x = np.arange(len(y))

# plot with various axes scales
plt.figure()

# linear
plt.subplot(221)
plt.plot(x, y)
plt.yscale('linear')
plt.title('linear')
plt.grid(True)

# log
plt.subplot(222)
plt.plot(x, y)
plt.yscale('log')
plt.title('log')
plt.grid(True)

# symmetric log
plt.subplot(223)
plt.plot(x, y - y.mean())
plt.yscale('symlog', linthresh=0.01)
plt.title('symlog')
plt.grid(True)

# logit
plt.subplot(224)
plt.plot(x, y)
plt.yscale('logit')
plt.title('logit')
plt.grid(True)
# Adjust the subplot layout, because the logit one may take more space
# than usual, due to y-tick labels like "1 - 10^{-3}"
plt.subplots_adjust(top=0.92, bottom=0.08, left=0.10, right=0.95, hspace=0.25,
                    wspace=0.35)

plt.show()
```

```python
# Building up a chart on explicit interface utlizing instances of axes.Axes to build up the visulatisation in an instance of figure.Figure
# The data contains sales information for a number of companies.
data = {'Barton LLC': 109438.50,
        'Frami, Hills and Schmidt': 103569.59,
        'Fritsch, Russel and Anderson': 112214.71,
        'Jerde-Hilpert': 112591.43,
        'Keeling LLC': 100934.30,
        'Koepp Ltd': 103660.54,
        'Kulas Inc': 137351.96,
        'Trantow-Barrows': 123381.38,
        'White-Trantow': 135841.99,
        'Will LLC': 104437.60}
group_data = list(data.values())
group_names = list(data.keys())
group_mean = np.mean(group_data)
```

```python
# This data is naturally visualized as a barplot, with one bar per group.
# To do this with the object-oriented approach, we first generate an instance of figure.Figure and axes.Axes.
# The Figure is like a canvas, and the Axes is a part of that canvas on which we will make a particular visualization.
fig, ax = plt.subplots()
```

```python
# We now have an Axes instance; we can plot on top of it
fig, ax = plt.subplots()
ax.barh(group_names, group_data)
```

```python
# Controlling the style
plt.style.available # list of styles available
plt.style.use('fivethirtyeight') # activates a style
fig, ax = plt.subplots()
ax.barh(group_names, group_data)
```

```python
# Customizing the plot
plt.rcParams.update({'figure.autolayout': True}) # automatically make room for elements in the figures that we create
# fig, ax = plt.subplots(figsize=(8, 4)) # manually abjust the figure size (width, height)
fig, ax = plt.subplots()
ax.barh(group_names, group_data)
labels = ax.get_xticklabels() # get access to a-axis labels
plt.setp(labels, rotation=45, horizontalalignment='right') # set style of x-axis labels
ax.set(xlim=[-10000, 140000], xlabel='Total Revenue', ylabel='Company', title='Company Revenue') # add labels to the plot with the Artist.set() method
plt.show()
```

```python
# For labels, we can specify custom formatting guidelines in the form of functions
# define a function that takes an integer as input, and returns a string as an output
# used with Axis.set_major_formatter or Axis.set_minor_formatter, they will automatically create and use a ticker.FuncFormatter class
# For this function, x is the original tick label and pos is the tick position
def currency(x, pos):
    """The two arguments are the value and tick position"""
    if x >= 1e6: # check if >= to 1million
        s = f'${x*1e-6:1.1f}M' # apply millions formatting
    else:
        s = f'${x*1e-3:1.0f}K' # aplly thousands formatting
    return s
fig, ax = plt.subplots(figsize=(6, 8))
ax.barh(group_names, group_data)
labels = ax.get_xticklabels()
plt.setp(labels, rotation=45, horizontalalignment='right')
ax.set(xlim=[-10000, 140000], xlabel='Total Revenue', ylabel='Company',
       title='Company Revenue')
ax.xaxis.set_major_formatter(currency) # call currency formatter on x-axis and apply the x labels
```

```python
# Combining multiple visualizations
# It is possible to draw multiple plot elements on the same instance of axes.Axes
# To do this we call another one of the plot methods on that axes object
fig, ax = plt.subplots(figsize=(8, 8))
ax.barh(group_names, group_data)
labels = ax.get_xticklabels()
plt.setp(labels, rotation=45, horizontalalignment='right')
ax.axvline(group_mean, ls='--', color='r') # Add a vertical line based on the mean, here we set the style in the function call
for group in [3, 5, 8]: # Annotate new companies
    ax.text(145000, group, "New Company", fontsize=10,
            verticalalignment="center")
ax.title.set(y=1.05) # Now we move our title up since it's getting a little cramped
ax.set(xlim=[-10000, 140000], xlabel='Total Revenue', ylabel='Company',
       title='Company Revenue')
ax.xaxis.set_major_formatter(currency)
ax.set_xticks([0, 25e3, 50e3, 75e3, 100e3, 125e3]) # add ticks to x-axis
fig.subplots_adjust(right=.1) # adjusts the right margin of the subplots, providing more space on the right side of the figure
plt.show()
```
