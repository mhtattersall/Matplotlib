```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from scipy.stats import uniform
from scipy.stats import bernoulli
```

```python
# Example: (create a checkerboard)
z = np.zeros((8,8), dtype=int)
z[1::2, ::2] = 1 # slice each row starting at 1, stride 2 and assign the values to 1
z[::2, 1::2] = 1 # slice each col starting at 1, stride 2 and assign values to 1
sns.heatmap(z,annot=True,linewidths=5,cbar=False)
```

```python
# Using Matplotlib
fig,ax = plt.subplots() # returns 2 objects, fig (holder) as distinct from the ax (chart)
ax.plot(np.random.rand(20),label='A',c='g',marker='o',ls='-',markersize=5,linewidth=3) # this creates the chart. Label gives the line a label, c is colour, ls is linestyle
ax.plot(np.random.rand(20),label='B',c='r',marker='x',ls='-.',markersize=5,linewidth=3)
ax.plot(np.random.rand(20),label='C',c='b',marker='^',ls='--',markersize=5,linewidth=3)
ax.set_title('First Graph')
plt.legend(loc=1) #loc is the location of the legend
plt.show()
```

```python
n_rows = 2
n_cols = 2
colours=['r','g','y','b']
labels=['A','B','C','D']
fig,axes = plt.subplots(n_rows,n_cols) # create 4 subplots
count=0
for row_num in range(n_rows):
  for col_num in range(n_cols):
    ax = axes[row_num][col_num]
    ax.plot(np.random.rand(20),label=labels[count],c=colours[count])
    ax.set_title(f'Plot ({row_num+1}, {col_num+1})')
    count += 1
    ax.legend()
fig.suptitle('Main title')
fig.tight_layout()
plt.show()
```

```python
# Line plot - for a series of data points
fig = plt.figure()
ax = plt.axes()
x = np.linspace(0,10,1000)
ax.plot(x,np.sin(x))
```

```python
# Multiple line plot
t = np.arange(0., 5., 0.2)
plt.plot(t, t, t, t**2, t, t**4) # note you need to list multiple lines as pairs
```

```python
# Bar plot - relationship between a numeric and a categorical value
# Each categorical value has a bar and its size represents its numercial value
courses = ['Data Science',' Python','R','Statistics','Machine Learning']
students = [30,45,8,25,38]
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.bar(courses,students)
plt.xlabel("Courses")
plt.ylabel("Student Enrollment")
plt.title("Student Enrollment by Course")
```

```python
# Bar plot example using xticks
bars = ('Europe', 'Asia', 'North America', 'Australia', 'Africa')
y_pos = np.arange(len(bars))
stu = [16, 21, 50, 1, 1]
plt.bar(y_pos, stu, color=(0.2, 0.4, 0.6, 0.6))
plt.title("Student background")
plt.xticks(y_pos, bars)
```

```python
# Scatter plot - show relationship between 2 different data dimensions
# If both increase at the same time, they have a positive relationship.
# If one increases as the other decerases, they have a negative relationship.
# If there is no pattern, they don't have any relationship
N = 500
x = np.random.rand(N)
y = np.random.rand(N)
plt.scatter(x,y,c='g',alpha=1) #c='g': The color of the markers # alpha=1: The transparency value of the markers (1 is fully opaque)
plt.title('Scatter plot using Matplotlib')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
```

```python
# Histogram - show frequency of data in successive intervals of equal size
# Most commonly, the independent variable is plotted on the horizontal axis and the dependent variable on the vertical axis.
x = [21,22,23,4,5,6,77,8,9,10,31,32,33,34,35,36,37,18,49,50,100]
num_bins = 5
plt.hist(x,num_bins,facecolor='blue',alpha=0.5,density=False) #density=True shows probability
plt.title('Simple Histogram using Matplotlib')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```
